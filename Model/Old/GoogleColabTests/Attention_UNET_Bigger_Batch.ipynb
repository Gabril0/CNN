```python
#data augmentation

from PIL import Image
import random
from scipy import ndimage

def preprocess_mask(mask):
    if mask.dtype == np.float32:
        mask = (mask * 255).astype(np.uint8)

    if mask.shape[-1] != 1:
        mask = np.array(Image.fromarray(mask).convert('L'))

    threshold = 127
    mask = (mask > threshold).astype(np.uint8) * 255

    mask = mask / 255.0

    if len(mask.shape) == 2:
        mask = np.expand_dims(mask, axis=-1)

    return mask

def flip_horizontal(image_array, mask_array):
    aug_images = []
    aug_masks = []
    for image, mask in zip(image_array, mask_array):
        aug_image = np.flip(image, axis=1)
        aug_mask = np.flip(mask, axis=1)

        aug_images.append(aug_image)
        aug_masks.append(aug_mask)

    return aug_images, aug_masks

def flip_vertical(image_array, mask_array):
    aug_images = []
    aug_masks = []
    for image, mask in zip(image_array, mask_array):
        aug_image = np.flip(image, axis=0)
        aug_mask = np.flip(mask, axis=0)

        aug_images.append(aug_image)
        aug_masks.append(aug_mask)

    return aug_images, aug_masks

def random_rotation(image_array, mask_array):
    aug_images = []
    aug_masks = []
    for image, mask in zip(image_array, mask_array):
        random_number = random.randint(0, 359)

        img_rotated = ndimage.rotate(image, random_number, reshape=False)
        mask_rotated = ndimage.rotate(mask, random_number, reshape=False)

        height, width, _ = image.shape
        crop_height = int(height * 0.8)
        crop_width = int(width * 0.8)
        img_cropped = img_rotated[(height - crop_height) // 2: (height + crop_height) // 2,
                                  (width - crop_width) // 2: (width + crop_width) // 2]
        mask_cropped = mask_rotated[(height - crop_height) // 2: (height + crop_height) // 2,
                                    (width - crop_width) // 2: (width + crop_width) // 2]

        img_resized = cv2.resize(img_cropped, (width, height), interpolation=cv2.INTER_AREA)
        mask_resized = cv2.resize(mask_cropped, (width, height), interpolation=cv2.INTER_AREA)

        aug_images.append(img_resized)
        aug_masks.append(mask_resized)

    return aug_images, aug_masks

def optical_distortion(image_array, mask_array, strength=10, zoom=3):
    aug_images = []
    aug_masks = []
    for image, mask in zip(image_array, mask_array):

      center_x, center_y = width // 2, height // 2

      x_coords, y_coords = np.meshgrid(np.arange(width), np.arange(height))
      x_coords = x_coords - center_x
      y_coords = y_coords - center_y

      r = np.sqrt(x_coords**2 + y_coords**2)

      r_distorted = r * (1 + strength * r**2 / (width**2 + height**2)) / (1 + strength * zoom * r**2 / (width**2 + height**2))
      theta_distorted = np.arctan2(y_coords, x_coords)

      x_distorted = r_distorted * np.cos(theta_distorted) + center_x
      y_distorted = r_distorted * np.sin(theta_distorted) + center_y
      distorted_image = cv2.remap(image, x_distorted.astype(np.float32), y_distorted.astype(np.float32), cv2.INTER_LINEAR)
      distorted_mask = cv2.remap(mask, x_distorted.astype(np.float32), y_distorted.astype(np.float32), cv2.INTER_LINEAR)

      aug_images.append(distorted_image)
      aug_masks.append(distorted_mask)

    return aug_images, aug_masks

def grid_distortion(image_array, mask_array, rows=6, cols=6):
    aug_images = []
    aug_masks = []
    for image, mask in zip(image_array, mask_array):
        distort_grid = np.indices(image.shape[:2], dtype=np.float32)
        distort_grid += np.random.normal(size=distort_grid.shape) * 10

        distort_grid[0] = np.clip(distort_grid[0], 0, image.shape[0] - 1)
        distort_grid[1] = np.clip(distort_grid[1], 0, image.shape[1] - 1)

        map_x = distort_grid[1]
        map_y = distort_grid[0]

        distorted_image = cv2.remap(image, map_x, map_y, cv2.INTER_LINEAR)
        distorted_mask = cv2.remap(mask, map_x, map_y, cv2.INTER_LINEAR)

        aug_images.append(distorted_image)
        aug_masks.append(distorted_mask)

    return aug_images, aug_masks

def elastic_transformation(image_array, mask_array):
    aug_images = []
    aug_masks = []
    for image, mask in zip(image_array, mask_array):
        alpha = 50
        sigma = 5

        rows, cols, _ = image.shape

        random_field_x = alpha * cv2.GaussianBlur((np.random.rand(rows, cols) * 2 - 1), (sigma * 2 + 1, sigma * 2 + 1), 0)
        random_field_y = alpha * cv2.GaussianBlur((np.random.rand(rows, cols) * 2 - 1), (sigma * 2 + 1, sigma * 2 + 1), 0)

        grid_x, grid_y = np.meshgrid(np.arange(cols), np.arange(rows))
        grid_x = np.clip(grid_x + random_field_x, 0, cols - 1)
        grid_y = np.clip(grid_y + random_field_y, 0, rows - 1)

        elastic_image = cv2.remap(image, grid_x.astype(np.float32), grid_y.astype(np.float32), interpolation=cv2.INTER_LINEAR)
        elastic_mask = cv2.remap(mask, grid_x.astype(np.float32), grid_y.astype(np.float32), interpolation=cv2.INTER_LINEAR)

        aug_images.append(elastic_image)
        aug_masks.append(elastic_mask)

    return aug_images, aug_masks


augmented_images, augmented_masks = training_array, mask_array

all_augmented_images = [augmented_images]
all_augmented_masks = [augmented_masks]

im, ma = flip_horizontal(training_array, mask_array)
all_augmented_images.append(im)
all_augmented_masks.append(ma)
print("done with horizontal flip, total of images: " + str(len(all_augmented_images[-1])))

im, ma = flip_vertical(training_array, mask_array)
all_augmented_images.append(im)
all_augmented_masks.append(ma)
print("done with vertical flip, total of images: " + str(len(all_augmented_images[-1])))

im, ma = random_rotation(training_array, mask_array)
all_augmented_images.append(im)
all_augmented_masks.append(ma)
print("done with random rotation, total of images: " + str(len(all_augmented_images[-1])))

im, ma = grid_distortion(training_array, mask_array)
all_augmented_images.append(im)
all_augmented_masks.append(ma)
print("done with grid distortion, total of images: " + str(len(all_augmented_images[-1])))

im, ma = optical_distortion(training_array, mask_array)
all_augmented_images.append(im)
all_augmented_masks.append(ma)
print("done with optical distortion, total of images: " + str(len(all_augmented_images[-1])))

im, ma = elastic_transformation(training_array, mask_array)
all_augmented_images.append(im)
all_augmented_masks.append(ma)
print("done with elastic transformation, total of images: " + str(len(all_augmented_images[-1])))

#adding more rotations to training
for i in range(len(all_augmented_images)):
    im, ma = random_rotation(all_augmented_images[i], all_augmented_masks[i])
    all_augmented_images.append(im)
    all_augmented_masks.append(ma)
    print(f"Done with random rotation for image {i+1}, total of images: {len(all_augmented_images[-1])}")

augmented_images = [image for sublist in all_augmented_images for image in sublist]
augmented_masks = [mask for sublist in all_augmented_masks for mask in sublist]

preprocessed_augmented_images = augmented_images
preprocessed_test_array = test_array

preprocessed_test_masks = [preprocess_mask(mask) for mask in test_mask_array]
preprocessed_augmented_masks = [preprocess_mask(mask) for mask in augmented_masks]

```

```python
#constructing the UNET functions

def conv_block(inputs,filters):
  #two conv layers with batch normalization and relu, this makes features more apparent as the tiny weight get smaller and big ones grow
  conv = Conv2D(filters, kernel_size=(3, 3), strides=(1, 1), padding='same')(inputs)
  conv = BatchNormalization()(conv)
  conv = Activation('relu')(conv)
  conv = Conv2D(filters, kernel_size=(3, 3), strides=(1, 1), padding='same')(conv)
  conv = BatchNormalization()(conv)
  conv = Activation('relu')(conv)
  return conv

def attention_gate(input_tensor, skip_tensor, filters):
  #relu
  attn = Conv2D(filters, kernel_size=(1, 1), strides=(1, 1), padding='same')(input_tensor)
  attn = BatchNormalization()(attn)
  attn = Activation('relu')(attn)

  #sigmoid
  attn = Conv2D(filters, kernel_size=(1, 1), strides=(2, 2), padding='same')(attn)
  attn = BatchNormalization()(attn)
  attn = Activation('sigmoid')(attn)

  #resampler
  attn = UpSampling2D(size=(2, 2))(attn)
  attn = Conv2D(filters, kernel_size=(3, 3), strides=(1, 1), padding='same')(attn)

  #multiplication with the skip_tensor layer
  multiplied = tf.multiply(skip_tensor, attn)

  concatenated = Concatenate()([input_tensor, multiplied])
  return concatenated

def unet_build(input_shape):
  inputs = Input(input_shape)

  #downsampling
  conv1 = conv_block(inputs, 64)
  pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

  conv2 = conv_block(pool1, 128)
  pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

  conv3 = conv_block(pool2, 256)
  pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)

  conv4 = conv_block(pool3, 512)
  pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)

  #bridge
  conv5 = conv_block(pool4, 1024)

  #upsampling
  up6 = UpSampling2D(size=(2, 2))(conv5)
  up6 = attention_gate(up6, conv4, 512)
  conv6 = conv_block(up6, 512)

  up7 = UpSampling2D(size=(2, 2))(conv6)
  up7 = attention_gate(up7, conv3, 256)
  conv7 = conv_block(up7, 256)

  up8 = UpSampling2D(size=(2, 2))(conv7)
  up8 = attention_gate(up8, conv2, 128)
  conv8 = conv_block(up8, 128)

  up9 = UpSampling2D(size=(2, 2))(conv8)
  up9 = attention_gate(up9, conv1, 64)
  conv9 = conv_block(up9, 64)

  outputs = Conv2D(1, kernel_size=(1, 1), strides=(1, 1), activation='sigmoid')(conv9)

  model = Model(inputs=inputs, outputs=outputs)
  return model
```
